Seaborn --> It is a library is used for Data Visualize


```python
#importing the library
import seaborn as syn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Note: Seaborn has some built-in datasets


```python
tip=syn.load_dataset('tips')
```

```python
tip.head()
```

```python
tip.tail()
```

```python
#set theme
syn.set_theme()
```

```python
#visualize the data
syn.relplot(data=tip, x='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
#load the iris dataset
iris=syn.load_dataset('iris')
```

```python
iris.head()
```

```python
syn.relplot(data=iris,x='sepal_length',y='sepal_width',hue='petal_length',style='species',col='species')
```

Scatter plot

```python
syn.scatterplot(data=iris,x='sepal_length',y='petal_width',hue='species')
```

```python
#loading the titanic dataset
titanic=syn.load_dataset('titanic')
```

```python
titanic.head()
```

```python
syn.scatterplot(data=titanic,x='age',y='fare',hue='embark_town',style='deck',size='alive')
```

```python
#theme of dataset
syn.set_theme(style='darkgrid')
```

Count plot

```python
syn.countplot(x='class',data=titanic)
```

```python
syn.countplot(x='class',data=titanic)
```

```python
titanic.shape
```

Bar plot

```python
syn.barplot(x='sex',y='survived',hue='class',data=titanic)
```

Distribution plot


```python
syn.displot(data=titanic,x='age',hue='sex')
```
